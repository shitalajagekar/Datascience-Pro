```r
duplicates <- c(4, 6, 11, 13, 15, 17, 18, 20, 22, 24, 25, 28, 32, 34, 36, 38, 40, 44, 46, 48, 51, 54, 65, 158)
```

```r
food2 <- food[, (-duplicates)]
```

```r
dim(food2)
```

```r
useless <- c(1:3, 32:41)
food3 <- food2[, (-useless)]
```

```r
library(stringr)
```

```r
nutrition <- str_detect(names(food3), "100g")
```

```r
summary(food3[,nutrition])
```

```r
food3$sugars_100g[is.na(food3$sugars_100g)] <- 0
```

```r
hist(food3$sugars_100g, breaks = 100)
```

```r
food4 <- food3[food3$sugars_100g > 0, ]
```
